```python
with open("../data/brown-universal.txt") as f:
    data = f.readlines()
```

```python
##todo:  词性之间的状态转移矩阵
##todo:  加入每句话的起始和终止
```

```python
data = [j for j in [i.strip().lower().split("\t") for i in data if i != "\n"] if len(j) == 2]
```

```python
data[:14]
```

```python
dwt = {}
dtw = {}
for pair in data:
    if pair[0] not in dwt:
        dwt[pair[0]] = {}
    if pair[1] not in dwt[pair[0]]:
        dwt[pair[0]][pair[1]] = 1
    else:
        dwt[pair[0]][pair[1]] += 1
        
    if pair[1] not in dtw:
        dtw[pair[1]] = {}
    if pair[0] not in dtw[pair[1]]:
        dtw[pair[1]][pair[0]] = 1
    else:
        dtw[pair[1]][pair[0]] += 1
        
        
```

```python
dw = dict([(word, sorted(ts.items(), key= lambda x: -x[1])[0][0]) for word, ts in dwt.items()])
```

```python
dt = {}
for t in dtw.keys():
    dt[t] = {}
    s = sum(dtw[t].values())
    for n in dtw[t].keys():
        dt[t][n] = dtw[t][n] / s
```
