```python
import matplotlib.pyplot as plt
%matplotlib inline
from random import random
import numpy as np
from numpy import sin, cos, pi
```

```python
def pointGenerator(a):
    r = a + random()
    theta = random()*2*pi
    return [r*cos(theta),  r*sin(theta)]
data = np.array([pointGenerator(1) for _ in range(300)] + [pointGenerator(3) for _ in range(500)])
plt.scatter(data[:,0], data[:,1])
```
