```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def box_muller():
    u1 = random.uniform(0,1)
    u2 = random.uniform(0,1)
    z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2) 
    z1 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    return z0, z1
```

```python
plt.hist([box_muller()[0] for _ in range(234000)], bins=150)
plt.show()
```

```python
plt.hist([box_muller()[1] for _ in range(234000)], bins=150)
plt.show()
```
