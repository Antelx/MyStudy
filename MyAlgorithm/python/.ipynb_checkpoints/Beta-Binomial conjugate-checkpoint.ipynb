```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
n = 10
k = 7
```

```python
def gen(n, k, count= 20):
    a = k
    b = n - k + 1
    prior = np.random.beta(a, b)
    m1 = sum([np.random.binomial(1, prior) for _ in range(count)])
    m2 = count - m1
    post = np.random.beta(a+m1, b+m2)
    return (prior, post, m1, m2)
```

```python
d = [gen(n, k) for _ in range(200000)]
```

```python
plt.hist([i[0] for i in d], bins=100, alpha=0.3)
plt.hist([i[1] for i in d], bins=100, alpha=0.3)
plt.show()
```

```python
def gen2():
    return sorted([np.random.uniform() for _ in range(n)])[k-1]    

d2 = [gen2() for _ in range(200000)]
plt.hist([i for i in d2], bins=100, alpha=0.3)
plt.show()
```

```python
plt.hist([np.random.beta(k, n-k+1) for _ in range(200000)], bins=100, alpha=0.3)
plt.show()
```

```python
def gen3(n, k, count= 20):
    a = k 
    b = n - k + 1
    
    target = sorted([np.random.uniform() for _ in range(n)])[k-1]    
    
    m1 = sum([np.random.binomial(1, target) for _ in range(count)])
    m2 = count - m1
    
    post = np.random.beta(a+m1, b+m2)
    return post

d3 = [gen3(n, k) for _ in range(200000)]
plt.hist([i for i in d3], bins=100, alpha=0.3)
plt.show()
```
