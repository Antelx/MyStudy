```python
r"""
指定一步转移概率和初始状态
得到马尔科夫链每一步的状态
"""
```

```python
import numpy as np
```

```python
P = np.matrix([
    [0.65, 0.28, 0.07],
    [0.15, 0.67, 0.18],
    [0.12, 0.36, 0.52]
])
```

```python
pi0 = np.matrix([0.7, 0.2, 0.1])
```

```python
pi = pi0
for _ in range(30):
    print(pi)
    pi = np.matmul(pi, P)
```

```python
pi0 = np.matrix([0.17, 0.23, 0.6])
```

```python
pi = pi0
for _ in range(30):
    print(pi)
    pi = np.matmul(pi, P)
```
